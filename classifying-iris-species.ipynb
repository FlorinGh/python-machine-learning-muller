# Classifying Iris Species

### Challenge

In this section we will create a simple machine learning application for classify iris flowers.  
We have a collection of measurements associated with iris flowwers: the length and width of the petals and length and width of the sepals, in centimeters.  
There are three classes available: setosa, versicolor and virginica.

**Because we have measurements for which we know the correct species of iris, this is a supervised learning problem. In this problem we want to predict one of several options. This is a classification problem.** The possible outputs are called *classes*; the known class for particular data point is called *label*.

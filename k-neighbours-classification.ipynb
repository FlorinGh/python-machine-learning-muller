# k-Neighbours Classification

The k-NN algorithm us arguably the simplest machine learning algorithm. Building the model consists only of storing the training dataset. To make a prediction for a new data point, the algorithm finds the closest data points in the training set: its nearest neighbours.

```python
from preamble import *
%matplotlib inline
```

The simples knn considers only the closest data point and assigns that class to it.

```python
mglearn.plots.plot_knn_classification(n_neighbors=1)
```

When using more neighbors, a voting system is used to assign the label; the algorithm finds the closest k neighbors and counts the number for each class; wichever class has the majority defines the label.

```python
mglearn.plots.plot_knn_classification(n_neighbors=3)
```

Using scikit-learn to train a knn model on the forge dataset:

```python
from sklearn.model_selection import train_test_split
X, y = mglearn.datasets.make_forge()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)
```

```python
clf.fit(X_train, y_train)
```

```python
print("Test set predictions: {}".format(clf.predict(X_test)))
```

```python
print("Test set accuracy: {:.3f}".format(clf.score(X_test, y_test)))
```

### Visualising the decision boundary

```python
fig, axes = plt.subplots(1, 5, figsize=(17, 3))

for n_neighbors, ax in zip([1, 3, 9, 25, 26], axes):
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.5, ax=ax, alpha=0.4)
    mglearn.discrete_scatter(X[:,0], X[:,1], y, ax=ax)
    ax.set_title("{} neighbor(s)".format(n_neighbors))
    ax.set_xlabel("feature 0")
    ax.set_ylabel("feature 1")
```

There are a few take aways from these plots:
* when we use only one neighbor the decision boundary follows closely the data set
* the boundary becomes smoother as we increase the number of neighbors
* a smooth decision boundary coresponds to a simpler model, one that generalizes better the dataset
* the maximum number of neighbors must have a limit as if we consider the entire dataset all new cases wil have the same prediction
* we will see next how accuracy varies with the neighbors number

### The Cancer Dataset

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify = cancer.target, random_state=23)

training_accuracy = []
test_accuracy = []

neighbors_settings = range(1, 426)

for n_neighbors in neighbors_settings:
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    training_accuracy.append(clf.score(X_train, y_train))
    test_accuracy.append(clf.score(X_test, y_test))

plt.figure(figsize=(16,5))
plt.plot(neighbors_settings, training_accuracy, label="training accuracy")
plt.plot(neighbors_settings, test_accuracy, label="test accuracy")
plt.xlabel("n_neighbors")
plt.ylabel("Accuracy")
plt.legend()
```

A few notes on the graph above:  
* we can see that the training accuracy starts from maximum, 1; this is because using only one neighbor provides the maximum complexity model which fits perfectly the training set; the fact that this is actually an overfit is proven by the accuracy of the test data set for 1 neighbour: the graph show we have a local minimum
* as we increase the number of neighbors the training accuracy decreases sharply; after 310 neighbors this value reaches a flat 60% accuracy
* a similar trend is noticed for the test accuracy except that we can clearly see a global maximum; this is in fact the best model for this algorithm
* plotting the first 11 values of each list we can determine the best n_neighbor for the model: n_neighbor=3, test_accuracy=0.965

```python
plt.figure(figsize=(10,5))
plt.plot(neighbors_settings[:11], training_accuracy[:11], label="training accuracy")
plt.plot(neighbors_settings[:11], test_accuracy[:11], label="test accuracy")
plt.xlabel("n_neighbors")
plt.ylabel("Accuracy")
plt.legend()
```

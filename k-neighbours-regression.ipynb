### Using the kNN Regressor

```python
from sklearn.neighbors import KNeighborsRegressor

X, y = mglearn.datasets.make_wave(n_samples=40)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

reg = KNeighborsRegressor(n_neighbors=3)
reg.fit(X_train, y_train)

print("Test set predictions: \n {}".format(reg.predict(X_test)))
```

We can evaluate the model using the score method; for regressors this returns the R^2 score (**coefficient of determination**).  
The R^2 score is a measure of goodnessof a prediction for a regression model and yields a score between 0 and 1; A value of 1 corresponds to a perfect prediction; a value of 0 corresponds to a model that predicts the same value for all tests: the mean of the training set response (y_train). 

```python
print("Test set R^2 score: {:.2f}".format(reg.score(X_test, y_test)))
```

A value above 0.8 indicates a relatively good model fit.

### Analyzing the KNeighborsRegressor

```python
fig, axes = plt.subplots(1, 3, figsize=(15,4))

line = np.linspace(-3, 3, 1000).reshape(-1, 1)

for n_neighbors, ax in zip([1, 3, 9], axes):
    reg = KNeighborsRegressor(n_neighbors)
    reg.fit(X_train, y_train)
    
    ax.plot(line, reg.predict(line), c="lime")
    ax.plot(X_train, y_train, '^', c=mglearn.cm2(0), markersize=5)
    ax.plot(X_test, y_test, 'v', c=mglearn.cm2(1), markersize=5)
    
    ax.set_title("{} neighbors(s)\n train score: {:.2f} test score: {:.2f}"
                 .format(n_neighbors, reg.score(X_train, y_train), reg.score(X_test, y_test)))
    ax.set_xlabel("Feature")
    ax.set_ylabel("Target")
axes[0].legend(["Model prediction", "Training set", "Test set"], loc="best")
    
```

It is always recommended to output the score of the model; as it can be seen above, using only one neighbor overfits the training set; the score is perfect for the training set, but subjected to the test set the model fails miserably (score 0.35).  
Increasing the number of neighbors simplifies the model but after a certain value the model starts to underfit the data.  
Creating a visualisation such as this, or even just a simple graph of the two scores (training and test) can help determining the best number of beighbors for a good model. In this case, 3 is better than 1 and 9.
